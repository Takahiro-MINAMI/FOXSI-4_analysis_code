```python
import ROOT
ROOT.gStyle.SetOptStat(0)

# ROOTで出力を非表示にする（オプション）
ROOT.gROOT.SetBatch(True)



#no2022_02 Vth

# txtファイルからデータを読み込む
file_path = 'cal_func/no2022_02/ped_sigma_adc.txt'
data = []
counter=0
with open(file_path, 'r') as file:
    for line in file:
        # 各行のデータをfloatに変換し、dataリストに追加
        data.append(float(line.strip()))
        counter+=1
        if(counter>128):
            break

# ヒストグラムの設定 (100ビン、範囲は1から5までに調整)
hist = ROOT.TH1F("histogram", "Data Histogram", 100, 0, 10)

# データをヒストグラムに追加
for value in data:
    hist.Fill(value)

# ヒストグラムを表示
canvas = ROOT.TCanvas("canvas", "Histogram Canvas", 800, 600)
hist.Draw()
hist.SetLineColor(ROOT.kBlack)
hist.SetLineWidth(2)
hist.GetXaxis().SetRangeUser(0,6)
hist.GetYaxis().SetRangeUser(0,55)
print(hist.GetMean())
# ヒストグラムを保存 (例: PNG形式)
#canvas.SaveAs("histogram.png")

# ヒストグラムを画面に表示する場合は以下を使用
canvas.Draw()
```
